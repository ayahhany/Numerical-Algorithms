## BFGS
- Implement the <b>BFGS</b> as a function and find the minimum of the following function:
    - $ğ’‡(ğ’™)=ğ’™^ğŸ+y^2$
    - $ğ’‡(ğ’™)=ğ’™^ğŸ+xy+y^2$  
    - $ğ’‡(ğ’™)=ğ’™^ğŸy$ 
- The stop condition is grad < epsilon.
- Try different epsilons and initial conditions.
- Compare the results from Gradient descent and Newton's method.

```python
import numpy as np
```

```python
def bfgs(alpha,x1,x0, eps):
    d = len(x0) # dimensions of the problem
    A0 = np.asarray(x0)
    A1 = np.asarray(x1)
    
    I = np.eye(d, dtype=int)
    B_inv = I
    for i in range(1000):
        DfA0 = np.array((Dfx(A0), Dfy(A0))) # initial gradient
        DfA1 = np.array((Dfx(A1), Dfy(A1))) # initial gradient

        print(f'xn_1: {A0} \nxn: {A1}')
        print(f'Dxn_1: {np.array([Dfx(A0), Dfy(A0)])} \nDxn: {np.array([Dfx(A1), Dfy(A1)])}')
        print(np.linalg.norm(DfA1))
        print('*****')
        
        delta_x = A1 - A0
        delta_y = DfA1 - DfA0
        
        ro = 1.0 / (np.dot(delta_y, delta_x))
        q1 = I - ro * delta_x[:, np.newaxis] * delta_y[np.newaxis, :]
        q2 = I - ro * delta_y[:, np.newaxis] * delta_x[np.newaxis, :]
        B_inv = np.dot(q1, np.dot(B_inv, q2)) + (ro * delta_x[:, np.newaxis] *
                                           delta_x[np.newaxis, :])
        
        #B_inv = np.dot(I - ((delta_x*delta_y) / np.dot(delta_y, delta_x)), np.dot(B_inv, I - ((delta_y*delta_x) / np.dot(delta_y, delta_x)) )) + ((delta_x*delta_x) / np.dot(delta_y, delta_x))
        res = np.array([A1, DfA1]).reshape(-1,2)
        
        A0 = A1
        A1 = A1 - alpha * np.dot(B_inv, DfA1)
        
        if np.linalg.norm(DfA1) < eps:
            
            print('Converged!')
            break
        print(f'B_inv: {B_inv}')
        print('*****************')
    return i , res
```

```python
#first function
f = lambda A: A[0]**2 + A[1]**2
Dfx = lambda A: 2*A[0] 
Dfy = lambda A: 2*A[1]
DDfx = lambda A: 2
DDfy = lambda A: 2
DDfxy = lambda A: 0
DDfyx = lambda A: 0
```

```python
## First function
## BFGD with xk-1 = (10,10) and x0 = (9,9), epsilon = 0.001
i , res = bfgs(1,(9,9),(10,10), 0.001)
print(f'Found solution after {i} iterations.')
print(res)
```
