```python
from prettytable import PrettyTable
## first equation : ùíô^ùüë‚àíùíô^ùüê‚àíùüèùüìùíô+ùüè$, ùíô_ùüé=ùüí.ùüì
## f'(x) =  3x^2-x-15
table = PrettyTable(["Iter", "Pn", "Pn+1", "|Pn+1 - Pn|"])
k = 1
x= 4.5
while(1):
  gxk = x**3 - x**2 -15*x +1
  gxk1 = 3*x**2 - x - 15
  xnew = x - (gxk/gxk1)
  table.add_row([k, x, gxk, xnew])
  if(abs(x-xnew)<0.00001 or k==100): break
  x=xnew
  k=k+1
print(table)
print('r=',xnew)
print('f(r)=',xnew**3 - xnew**2 -15*xnew +1)
```
