```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-50, 65,50)
f = lambda x: x**2 - 15*x +1
Df = lambda x: 2*x -15
DDf = lambda x: 2
plt.plot(x,f(x))
plt.grid(True)
plt.show()
plt.plot(x,Df(x))
plt.grid(True)
plt.show()
plt.plot(x,[DDf(x)]*len(x))
plt.grid(True)
plt.show()
```

```python
def gradient(alpha, x0, eps):
    x = x0
    for i in range(1000):
        grad = Df(x)
        x = x - alpha * grad
        if Df(x) < eps:
            print('Converged!')
            break
    return i, x, Df(x)
```

```python
f = lambda x: x**2 - 15*x +1
Df = lambda x: 2*x -15
DDf = lambda x: 2
```

```python
# Gradient descent with learning rate = 0.01, x0=40, epsilon = 0.1
i , x , f = gradient(0.01, 40, 0.1)
print(f'Found solution after {i+1} iterations.')
print((x,f))
```

```python
def newton(alpha, x0, eps):
    x = x0
    for i in range(1000):
        x = x - alpha * ((Df(x) / DDf(x)))
        if Df(x) < eps:
            print('Converged!')
            break
    return i, x, Df(x)
```

```python
# Newton's method without learning rate, x0=40, epsilon = 0.1
i , x , f = newton(1, 40, 0.1)
print(f'Found solution after {i+1} iterations.')
print((x,f))
```

```python
# Newton's method with learning rate = 0.8, x0=40, epsilon = 0.1
i , x , f = newton(0.8, 40, 0.1)
print(f'Found solution after {i+1} iterations.')
print((x,f))
```
